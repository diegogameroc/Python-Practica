# Primeros pasos con Matplotlib


```python
1# Modulos
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Grafico 1 : Grafiquemos un conjunto de pares ordenados
# El objeto primerario para construir un grafico es un objeto de tipo lista
# Definamos los objetos que conformen cada uno de los ejes
x = [-0.3,2.4,3.9,4.5]
y = [-6,3,7,1]

# Diagrma de disporsion
plt.plot(x,y)
#Ejecutar
plt.show()
```

```python
# Grafico 2
#Usemos datos de tipo ndarray para definiar los elementos de cada uno de los ejes
# Regla de correspondencia: y = f(x)
# Definamos un dominio (conjunto de partida): Usando la funcion linspace

#del intervalo (-33,33) generar 250 puntos equiespaciados
x1 = np.linspace(-33,33,250)*np.pi/180

#Para definir los elementos que van en el eje Y utilizamos una regla de correspondencia
# Aprovechar que las operaciones matematicas entre elementos de tipo ndarray son
# vectorizadas (elemento a elemento)

y1= np.sin(x1**(1/3)*np.cos(x1**(2.71)))

# Diagrma de disporsion
plt.plot(x1,y1)
#Ejecutar
plt.show()

```

```python
# Agregar un ruido a la señal definida por {x1,y1} del grafico 2
x1 = np.linspace(-33,33,250)*np.pi/180
y2= y1 + np.random.random(x1.shape[0])*10**(-1)

# Mostremos el diagrma de dispersion
plt.plot(x1,y2)
#Ejecutar
plt.show()
```

```python
# Grafico 4 : Coloquemos dos conjuntos de pares ordenados en una misma ventana
# Definamos un conjunto de partida
x = np.linspace(5,25,300)

# Definamos dos reglas de correspondencia: Dos variables dependientes
z1 = np.exp(x**(1/7))
z2=np.exp(-x**(2.1))

#Mostremos los dos conjuntos de pares ordenados: (x,z1) y (x,z2)
plt.plot(x,z1,x,z2)
plt.show()
```

```python
# Grafico 6 : Dada una funcion R2. Calcular su derivada en un punto determinado

# Conjunto de partida
A = np.linspace(3, 4, 300)

# Regla de correspondencia de la funcion
B = np.sin(A**2 + 3*A - 0.9) + (np.cos(A))**3

# Trazamos la tangente en x = 3.62 en [3, 4, 3.8]
# derivada : (2*A-3)*(np.cos(A**2 - 3*A - 0.9)) - 3*np.sin(A)*(np.cos(A))**2
xp = 3.35
yp = np.sin(xp**2 + 3*xp - 0.9) + (np.cos(xp))**3
m = (2*xp + 3)*(np.cos(xp**2 + 3*xp - 0.9)) - 3*np.sin(xp)*(np.cos(xp))**2

# Para un punto de paso (xp, yp) la recta tangente : y - yp = m*(x - xp)
dom_derivada = np.linspace(xp-0.1,xp+0.1,50)
y_derivada = m * (dom_derivada - xp) + yp

# Mostrar el conjunto de pares ordenados (A, B)
plt.plot(A, B,"b", dom_derivada, y_derivada,"--g")

plt.show()
```
